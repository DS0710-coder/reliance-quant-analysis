```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import mplfinance as mpf

# Download 2 years of Reliance Industries data from NSE
data = yf.download('RELIANCE.NS', period='2y', auto_adjust=True)

# Preview first 5 rows
print(data.head())
```

```python
# Plot the closing price over 2 years
data['Close'].plot(figsize=(12, 5), title='Reliance Industries - Last 2 Years', color='blue')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.grid(True)
plt.show()
```

```python
# Calculate daily percentage returns
data['Returns'] = data['Close'].pct_change()

# Preview first 5 rows
print(data['Returns'].head())
```

```python
# Plot daily returns with zero line
data['Returns'].plot(figsize=(12, 5), title='Reliance Industries - Daily Returns', color='green')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.grid(True)
plt.show()
```

```python
# Find the single worst trading day
worst_day = data['Returns'].idxmin()
print('Worst day:', worst_day)
print('Return:', data['Returns'].loc[worst_day])

# Note: This was June 4 2024 - Indian Election Results Day
# Reliance dropped ~7.5% in a single day due to market panic
```

```python
# Calculate 20-day and 50-day moving averages
data['MA20'] = data['Close'].rolling(20).mean()
data['MA50'] = data['Close'].rolling(50).mean()

# Plot close price with both moving averages
data[['Close', 'MA20', 'MA50']].plot(
    figsize=(12, 5), 
    title='Reliance - Close Price with Moving Averages'
)
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.grid(True)
plt.show()
```

```python
# Generate trading signals based on MA crossover
# 1  = BUY  (MA20 above MA50 - uptrend)
# -1 = SELL (MA20 below MA50 - downtrend)
# 0  = NEUTRAL (not enough data yet)

data['Signal'] = 0
data.loc[data['MA20'] > data['MA50'], 'Signal'] = 1
data.loc[data['MA20'] < data['MA50'], 'Signal'] = -1

print(data['Signal'].value_counts())
```

```python
# Calculate strategy returns
# shift(1) is critical - we act on yesterday's signal at today's price
# This avoids lookahead bias (cheating with future data)
data['Strategy_Returns'] = data['Signal'].shift(1) * data['Returns']

# Calculate cumulative returns for both strategy and buy & hold
data['Cumulative_Market'] = (1 + data['Returns']).cumprod()
data['Cumulative_Strategy'] = (1 + data['Strategy_Returns']).cumprod()

# Plot comparison
data[['Cumulative_Market', 'Cumulative_Strategy']].plot(
    figsize=(12, 5),
    title='Strategy vs Buy & Hold'
)
plt.xlabel('Date')
plt.ylabel('Growth of Rs 1')
plt.grid(True)
plt.show()
```

```python
# Annualized return
strategy_annual = data['Strategy_Returns'].mean() * 252
market_annual = data['Returns'].mean() * 252

# Annualized volatility
strategy_vol = data['Strategy_Returns'].std() * np.sqrt(252)
market_vol = data['Returns'].std() * np.sqrt(252)

# Sharpe ratio (risk free rate ~6% in India)
risk_free = 0.06
strategy_sharpe = (strategy_annual - risk_free) / strategy_vol
market_sharpe = (market_annual - risk_free) / market_vol

# Max drawdown
cumulative = data['Cumulative_Strategy']
rolling_max = cumulative.cummax()
drawdown = (cumulative - rolling_max) / rolling_max
max_drawdown = drawdown.min()

print(f'--- STRATEGY ---')
print(f'Annual Return:  {strategy_annual:.2%}')
print(f'Volatility:     {strategy_vol:.2%}')
print(f'Sharpe Ratio:   {strategy_sharpe:.2f}')
print(f'Max Drawdown:   {max_drawdown:.2%}')
print(f'\n--- BUY & HOLD ---')
print(f'Annual Return:  {market_annual:.2%}')
print(f'Volatility:     {market_vol:.2%}')
print(f'Sharpe Ratio:   {market_sharpe:.2f}')
```

```python
plt.style.use('dark_background')

fig = plt.figure(figsize=(14, 14))
gs = gridspec.GridSpec(3, 1, hspace=0.4)

# Chart 1 - Price & Moving Averages
ax1 = fig.add_subplot(gs[0])
ax1.plot(data.index, data['Close'], color='#00b4d8', linewidth=1.2, label='Close')
ax1.plot(data.index, data['MA20'], color='#f77f00', linewidth=1, label='MA20')
ax1.plot(data.index, data['MA50'], color='#06d6a0', linewidth=1, label='MA50')
ax1.set_title('Reliance — Price & Moving Averages', color='white', fontsize=13, pad=12)
ax1.legend(facecolor='#1e1e1e', edgecolor='none', fontsize=9)
ax1.grid(color='#2a2a2a', linewidth=0.5)
ax1.set_facecolor('#0d0d0d')

# Chart 2 - Buy/Sell Signals
ax2 = fig.add_subplot(gs[1])
ax2.plot(data.index, data['Close'], color='#00b4d8', linewidth=1, label='Close')
buy_signals = data[data['Signal'] == 1]
sell_signals = data[data['Signal'] == -1]
ax2.scatter(buy_signals.index, buy_signals['Close'],
            marker='^', color='#06d6a0', s=60, label='BUY', zorder=5)
ax2.scatter(sell_signals.index, sell_signals['Close'],
            marker='v', color='#ef233c', s=60, label='SELL', zorder=5)
ax2.set_title('Buy & Sell Signals', color='white', fontsize=13, pad=12)
ax2.legend(facecolor='#1e1e1e', edgecolor='none', fontsize=9)
ax2.grid(color='#2a2a2a', linewidth=0.5)
ax2.set_facecolor('#0d0d0d')

# Chart 3 - Strategy vs Buy & Hold
ax3 = fig.add_subplot(gs[2])
ax3.plot(data.index, data['Cumulative_Strategy'],
         color='#06d6a0', linewidth=1.5, label='Strategy')
ax3.plot(data.index, data['Cumulative_Market'],
         color='#ef233c', linewidth=1.5, label='Buy & Hold')
ax3.axhline(y=1, color='white', linestyle='--', linewidth=0.7, alpha=0.4)
ax3.set_title('Strategy vs Buy & Hold', color='white', fontsize=13, pad=12)
ax3.legend(facecolor='#1e1e1e', edgecolor='none', fontsize=9)
ax3.grid(color='#2a2a2a', linewidth=0.5)
ax3.set_facecolor('#0d0d0d')

fig.patch.set_facecolor('#0d0d0d')
plt.show()
```
